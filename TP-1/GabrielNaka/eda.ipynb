```python
# Imports

import pandas
```

```python
# Carregando dados
# Foi removido manualmente a primeira linha de cada arquivo para possibilitar leitura de dados


Education = pandas.read_csv('Education.csv')
Expenditure = pandas.read_csv('Expenditure.csv')
Crime = pandas.read_csv('Crime.csv')
```

```python
# Verificando dados faltantes

print ("Dados faltantes por coluna: ")
Education.isnull().sum() + Expenditure.isnull().sum() + Crime.isnull().sum()

# Todos os dados faltantes se localizam em 'Footnotes'
# Ausencia nao afeta analise significativamente

# 'Unnamed: 1' e uma coluna vazia, sera ignorada
```

```python
# Adquirindo campos

camposEducation = set()
camposExpenditure = set()
camposCrime = set()


            # Iterando 'Series' por campos unicos

for line in Education.itertuples():
    camposEducation.add(line.Series)

for line in Expenditure.itertuples():
    camposExpenditure.add(line.Series)

for line in Crime.itertuples():
    camposCrime.add(line.Series)



            # Imprimindo campos

for index in range(len(camposEducation)):
    campo = list(camposEducation)[index]
    print(f"Education({index}): {campo}")

for index in range(len(camposExpenditure)):
    campo = list(camposExpenditure)[index]
    print(f"Expenditure({index}): {campo}")

for index in range(len(camposCrime)):
    campo = list(camposCrime)[index]
    print(f"Crime({index}): {campo}")
```

```python
# Adquirindo areas

areasDisponiveis = set()


            # Iterando 'Region/Country/Area' por campos unicos

for line in Education.itertuples():
    areasDisponiveis.add(line._2)  
    # '_2' == coluna 'Region/Country/Area', nao da pra colocar o nome certo por causa do '/'

for line in Expenditure.itertuples():
    areasDisponiveis.add(line._2)

for line in Crime.itertuples():
    areasDisponiveis.add(line._2)



            # Imprimindo campos

for index in range(len(areasDisponiveis)):
    area = list(areasDisponiveis)[index]
    print(f"Area({index}): {area}")
```
