```python

mt.plot(gasolina["Mes"], gasolina["Valor de Venda"], label="Gasolina")
mt.plot(petroleo["date"], petroleo["preco"],label="Barril  de petroleo")

mt.xlabel("DATA")
mt.ylabel("PREÇO")
mt.title("Crescimento do barril de petroleo vs valor na bomba")

mt.legend()
mt.show()
```

```python
#Comparamos o valor do do aumento do petroleo com o valor que estava na bomba para verificamros se os 2 crescem juntos
graph, x1 = mt.subplots()


x1.plot(gasolina["Mes"], gasolina["Valor de Venda"], label="Gasolina", color="red")
x1.set_ylabel("Gasolina", color="red")

x2 = x1.twinx()
x2.plot(petroleo["date"], petroleo["preco"],label="Barril  de petroleo")
x2.set_ylabel("Petróleo", color="blue")
mt.title("Crescimento do barril de petroleo vs valor na bomba")

mt.show()
```
