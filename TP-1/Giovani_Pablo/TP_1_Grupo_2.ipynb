## **Grupo 2: Limpeza e Preparação**

```python
# Grupo 2 Trabalho : Limpeza do datasetDados Nacionais de Segurança Pública - UF
#Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
!git clone https://github.com/claytontey/Aulas_Git-DS.git
```

```python
caminho_arquivo = './Aulas_Git-DS/TP-1/Giovani_Pablo/dados/indicadoressegurancapublicauf.xlsx'

# Importando a aba de Ocorrências
df_ocorrencias = pd.read_excel(caminho_arquivo, sheet_name='Ocorrências')

# Importando a aba de Vítimas
df_vitimas = pd.read_excel(caminho_arquivo, sheet_name='Vítimas')

# Verificando se a importação deu certo
display(df_ocorrencias.head(2))
display(df_vitimas.head(2))
```

```python
# 1. Verificando os dados nulos;
print("--- DADOS NULOS EM OCORRÊNCIAS ---")
display(df_ocorrencias.isnull().sum())

print("\n--- DADOS NULOS EM VÍTIMAS ---")
display(df_vitimas.isnull().sum())

# 2. Verificando os tipos de dados
print("\n--- TIPOS DE DADOS ---")
print("Ocorrências:\n", df_ocorrencias.dtypes)
print("\nVítimas:\n", df_vitimas.dtypes)
```

```python
# Função para limpar e padronizar colunas de texto
def padronizar_textos(df):
    for coluna in df.columns:
        # Verifica se a coluna é do tipo texto (object)
        if df[coluna].dtype == 'object':
            # 1. Remove espaços em branco no começo e no fim
            df[coluna] = df[coluna].str.strip()
            # 2. Deixa tudo em letras maiúsculas
            df[coluna] = df[coluna].str.upper()
    return df

# Aplicando a padronização nas duas bases
df_ocorrencias = padronizar_textos(df_ocorrencias)
df_vitimas = padronizar_textos(df_vitimas)
```

```python
# Verificando os valores únicos
print("--- VALORES ÚNICOS: OCORRÊNCIAS ---")
print("UF:", df_ocorrencias['UF'].unique())
print("Tipo Crime:", df_ocorrencias['Tipo Crime'].unique())
print("Mês:", df_ocorrencias['Mês'].unique())

print("\n--- VALORES ÚNICOS: VÍTIMAS ---")
print("Sexo da Vítima:", df_vitimas['Sexo da Vítima'].unique())
#Sem colunas duplicadas ou dados fantasmas
```

```python
# Verificando se existem linhas 100% idênticas repetidas
duplicadas_ocorrencias = df_ocorrencias.duplicated().sum()
duplicadas_vitimas = df_vitimas.duplicated().sum()

print(f"Linhas duplicadas em Ocorrências: {duplicadas_ocorrencias}")
print(f"Linhas duplicadas em Vítimas: {duplicadas_vitimas}")
```

```python
# Salvando os dados limpos em novos arquivos CSV
df_ocorrencias.to_csv('ocorrencias_limpo.csv', index=False, encoding='utf-8')
df_vitimas.to_csv('vitimas_limpo.csv', index=False, encoding='utf-8')
```
