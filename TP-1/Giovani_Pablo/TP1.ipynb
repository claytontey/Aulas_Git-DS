## Grupo 1: Aquisição de Dados

### 1. Título da Pesquisa
**Análise Temporal e Regional da Criminalidade no Brasil: Um Estudo sobre Ocorrências de Crimes contra o Patrimônio e a Vida (2019-2023).**

### 2. Dataset Escolhido
O conjunto de dados selecionado é o "Dados Nacionais de Segurança Pública - Unidade da Federação", disponível em formato .xlsx. Ele consolida as ocorrências criminais reportadas pelas Secretarias de Segurança Pública de todos os estados brasileiros.

### 3. Fonte dos Dados
Os dados são originários do SINESP (Sistema Nacional de Informações de Segurança Pública, Prisionais, de Rastreabilidade de Armas e Munições, de Material Genético, de Digitais e de Drogas).

Órgão Responsável: Ministério da Justiça e Segurança Pública (MJSP) / Secretaria Nacional de Segurança Pública (SENASP).

Acesso: Disponibilizado através do Portal de Dados Abertos do Governo Federal.

Confiabilidade: Trata-se de uma fonte oficial primária, cujos dados são alimentados mensalmente pelos estados brasileiros seguindo uma metodologia de padronização nacional.

### 4. Descrição do Problema Real
A segurança pública é um dos pilares da gestão governamental. No entanto, o Brasil possui dimensões continentais com realidades criminais muito distintas entre as regiões. O problema central reside na dificuldade de identificar padrões sazonais e disparidades regionais sem uma análise de dados robusta.

Para gestores, entender se um crime (como roubo de veículos) aumenta em determinados meses ou se concentra em estados específicos é crucial para a alocação inteligente de recursos, planejamento de operações e criação de políticas de prevenção baseadas em evidências (Data-Driven Policing).

### 5. Perguntas Analíticas
Para guiar o desenvolvimento do projeto e a futura análise preditiva/estatística, definimos as seguintes questões:

Sazonalidade: Existe um padrão mensal repetitivo para crimes contra o patrimônio (ex: Roubo a Instituição Financeira) ao longo dos anos analisados?

Disparidade Regional: Qual a variação percentual de crimes entre as Unidades da Federação e quais estados apresentam as maiores taxas de redução ou aumento em crimes violentos?

Correlação de Tipos Penais: Existe correlação estatística entre o aumento de furtos e o aumento de roubos em uma mesma região?

Evolução Histórica: Como a pandemia de COVID-19 (especialmente nos anos de 2020 e 2021) impactou o volume de ocorrências registradas em comparação aos anos pré e pós-isolamento social?
