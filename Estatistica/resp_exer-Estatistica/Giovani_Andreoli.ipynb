# 📊 Estudo de Caso: Desempenho de Alunos

Você foi contratado como cientista de dados para analisar o desempenho de uma turma.
Seu objetivo é gerar insights para melhorar o ensino.

Utilizando o material e as métricas apresentadas na aula de hoje, apresente seus _insigths_ para o _dataset_ apresentado:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_estudo_estatistica.csv')
df.head()
```

## Exercício 1
Descreva e analise algo relevante nos dados.

```python
import seaborn as sns
# Exercício 1: Distribuição Geral das Notas
print(f"Média de Notas: {df['nota'].mean():.2f}")
plt.figure(figsize=(8,4))
sns.histplot(df['nota'], kde=True, color='teal')
plt.title('Distribuição das Notas')
plt.show()
# Insight: A distribuição permite identificar se a maioria da turma está acima da média de aprovação.
```

## Exercício 2
Descreva e analise algo relevante nos dados.

```python
# Exercício 2: Correlação entre Horas de Estudo e Notas
correlation = df['horas_estudo'].corr(df['nota'])
print(f"Correlação: {correlation:.2f}")
sns.scatterplot(data=df, x='horas_estudo', y='nota')
plt.title('Horas de Estudo vs Nota Final')
plt.show()
# Insight: Quantifica o quanto o esforço individual (estudo fora da aula) impacta o resultado final.
```

## Exercício 3
Descreva e analise algo relevante nos dados.

```python
# Exercício 3: Impacto do Absenteísmo (Faltas)
plt.figure(figsize=(8,4))
sns.regplot(data=df, x='faltas', y='nota', color='red', scatter_kws={'alpha':0.3})
plt.title('Relação entre Faltas e Notas')
plt.show()
# Insight: Identifica se existe uma queda acentuada de desempenho conforme o número de faltas aumenta.
```

## Exercício 4
Descreva e analise algo relevante nos dados.

```python
# Exercício 4: Análise de Desempenho por Turno
display(df.groupby('turno')['nota'].describe())
sns.boxplot(data=df, x='turno', y='nota', palette='Set2')
plt.title('Comparativo de Notas: Manhã vs Noite')
plt.show()
# Insight: Verifica se o cansaço do turno da noite ou a rotina da manhã afeta a aprendizagem.
```

## Exercício 5
Descreva e analise algo relevante nos dados.

```python
# Exercício 5: Dificuldade por Linguagem de Programação
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='linguagem', y='nota', estimator=np.mean, errorbar=None)
plt.title('Nota Média por Linguagem')
plt.show()
# Insight: Ajuda a identificar linguagens que podem exigir uma metodologia de ensino diferente.
```

## Exercício 6
Descreva e analise algo relevante nos dados.

```python
# Exercício 6: Matriz de Correlação
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação de Variáveis')
plt.show()
# Insight: Visão sistêmica de quais fatores (acessos, tempo, faltas) mais se relacionam com a nota.
```

## Exercício 7
Descreva e analise algo relevante nos dados.

```python
# Exercício 7: Engajamento Digital (Acessos à Plataforma)
sns.histplot(df['acessos_plataforma'], bins=20, color='orange')
plt.title('Frequência de Acessos à Plataforma')
plt.show()
# Insight: Mostra se os alunos estão utilizando os recursos digitais disponibilizados.
```

## Exercício 8
Descreva e analise algo relevante nos dados.

```python
# Exercício 8: Relação Acessos vs Nota
sns.jointplot(data=df, x='acessos_plataforma', y='nota', kind='hex', color='blue')
plt.show()
# Insight: Verifica se o uso da plataforma se traduz em melhores notas.
```

## Exercício 9
Descreva e analise algo relevante nos dados.

```python
# Exercício 9: Perfil de Tempo de Resolução
print(f"Tempo Médio de Resolução: {df['tempo_resolucao'].mean():.2f} min")
sns.kdeplot(df['tempo_resolucao'], fill=True)
plt.title('Distribuição do Tempo de Resolução de Exercícios')
plt.show()
```

## Exercício 10
Descreva e analise algo relevante nos dados.
