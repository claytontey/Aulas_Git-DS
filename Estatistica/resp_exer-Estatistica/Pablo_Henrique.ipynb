## Exercício 13
Descreva e analise algo relevante nos dados.

```python
media_faltas = df.groupby('bolsa_estagio')['faltas'].mean()
mediana_faltas = df.groupby('bolsa_estagio')['faltas'].median()

print("Media de faltas por bolsa:")
print(media_faltas)

print("\nMediana de faltas por bolsa:")
print(mediana_faltas)

# visual
df.boxplot(column='faltas', by='bolsa_estagio')
plt.title('Faltas por Bolsa de Estagio')
plt.suptitle('')
plt.xlabel('Bolsa (0 = nao, 1 = sim)')
plt.ylabel('Faltas')
plt.show()

print("Buscando correlação entre as faltas e a presença de bolsas estudantis")
```

## Exercício 14
Descreva e analise algo relevante nos dados.

```python
q1 = df['nota'].quantile(0.25)
q3 = df['nota'].quantile(0.75)
iqr = q3 - q1
limite_inf = q1 - 1.5 * iqr
limite_sup = q3 + 1.5 * iqr

outliers = df[(df['nota'] < limite_inf) | (df['nota'] > limite_sup)]

print(f"Q1: {q1:.2f}")
print(f"Q3: {q3:.2f}")
print(f"IQR: {iqr:.2f}")
print(f"Limite inferior: {limite_inf:.2f}")
print(f"Limite superior: {limite_sup:.2f}")

print("\nPossiveis outliers:")
print(outliers[['nota']])

print(f"\nQuantidade de outliers: {outliers.shape[0]}")
```
