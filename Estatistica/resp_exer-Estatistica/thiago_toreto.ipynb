## Exercício 7
Descreva e analise algo relevante nos dados.

```python
# Qual é o coeficiente de variação da nota?
media = df['nota'].mean()
desvio_padrao = df['nota'].std()
cv = desvio_padrao / media
print(f"Coeficiente de variação: {cv:.4f}")
```

## Exercício 8

## Exercício 9
Descreva e analise algo relevante nos dados.

```python
# Qual linguagem tem a maior média de nota?
media_por_linguagem = df.groupby('linguagem')['nota'].mean()
melhor_linguagem = media_por_linguagem.idxmax()
maior_media = media_por_linguagem.max()
print(f"Linguagem com maior média: {melhor_linguagem}")
print(f"Média da nota: {maior_media:.2f}")
```
