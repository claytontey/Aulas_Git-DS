## Exercício 1
Descreva e analise algo relevante nos dados.

```python
media_horas_estudo = df["horas_estudo"].mean()
mediana_horas_estudo = df["horas_estudo"].median()
moda_horas_estudo = df["horas_estudo"].mode()
desvio_padrao =  df["horas_estudo"].std()
print("Média: ",media_horas_estudo)
print("Mediana: ", mediana_horas_estudo)
print("Moda: ",moda_horas_estudo)
print("Desvio padrão: ",desvio_padrao)
print("""\nANÁLISE\nA média de horas de estudo foi de aproximadamente 5,36 horas, com mediana de 5, indicando uma distribuição relativamente equilibrada dos dados. A moda igual a 8 horas sugere que um número significativo de alunos apresenta alta dedicação aos estudos. O desvio padrão de 2,62 indica uma variabilidade moderada, mostrando que, embora haja diferenças entre os alunos, a maioria se concentra em torno da média.""")
```

## Exercício 2
Descreva e analise algo relevante nos dados.

```python
faltas_turno = df.groupby('turno')['faltas']

print("\nMÉDIA\n",faltas_turno.mean())
print("\nMEDIANA\n",faltas_turno.median())
print("\nDESVIO PADRÃO\n",faltas_turno.std())
print("\nVARIÂNCIA\n",faltas_turno.var())
print("\nVALOR MÁXIMO\n",faltas_turno.max())
print("\nVALOR MÍNIMO\n",faltas_turno.min())
print("\nANÁLISE\nAqui é possível concluir que a média de faltas dos alunos que estudam no período da noite é maior que a do turno da manhã, notando que os alunos tendem a faltar mais no curso noturno, porém o desvio padrão é relativamente alto, indicando que há uma grande variabilidade, e que nem todos seguem esse padrão")
```

## Exercício 3
Descreva e analise algo relevante nos dados.

```python
correlacao = df[['faltas','nota']].corr()
media_faltas = df['faltas'].mean()
desvio = df['faltas'].std()
print(correlacao, "\nMédia de faltas: ",media_faltas, "\nDesvio: ",desvio, "\nANÁLISE\nAqui podemos notar que a relação entre esses dois atributos não é tão direta, mas que há uma pequena relação entre as faltas e as notas dos alunos, sendo quanto maior a quantidade de faltas, menores as notas. E é possível ver um alto grau de desvio padrão também, o que indica valores discrepantes e variados em relação à média das faltas, que era em torno de 5,275)")
```
