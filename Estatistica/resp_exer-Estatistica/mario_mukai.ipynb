## Exercício 1
Descreva e analise algo relevante nos dados.

```python
# resposta aqui
# média de horas de estudo dos alunos
media = df['horas_estudo'].mean()

print(f"A média de horas estudadas é {media:.2f} horas")
```

## Exercício 2
Descreva e analise algo relevante nos dados.

```python
# resposta aqui
# verificar a mediana para ver se a média condiz com a amostra

mediana = df['horas_estudo'].median()

print(f"A mediana de horas estudadas é {mediana:.2f} horas")

# assim podemos ver que a média esta ok
```

## Exercício 4

## Exercício 6

```python
# resposta aqui
# vamos ver se tem relação o acesso maior de noite com a quantidade de alunos que estudam de noite

df_qtd_aluno_noite = df[df['turno'] == 'Noite']

qtd_aluno_noite = len(df_qtd_aluno_noite)

print(f"A quantidade de alunos que estudam de noite é {qtd_aluno_noite} alunos")

df_qtd_aluno_manha = df[df['turno'] == 'Manhã']

qtd_aluno_manha = len(df_qtd_aluno_manha)

print(f"A quantidade de alunos que estudam de manhã é {qtd_aluno_manha} alunos")

# Assim vimos que condiz, ja que existem 10 aluno a mais de noite
```
