## Exercício 19
Descreva e analise algo relevante nos dados.

```python
# Relação de tendências entre horas de estudo e nota
x = df["horas_estudo"]
y = df["nota"]

coef = np.polyfit(x, y, 1)
polinomio = np.poly1d(coef)

plt.scatter(x, y, color="blue")
plt.plot(x, polinomio(x), color="red")
plt.title("Horas de Estudo vs Nota (com tendência)")
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota")
plt.show()
```

## Exercício 20
Descreva e analise algo relevante nos dados.

```python
# Nota média por horas de estudo

media_estudo = df.groupby("horas_estudo")["nota"].mean()

plt.plot(media_estudo.index, media_estudo.values, marker='o')
plt.title("Nota Média por Horas de Estudo")
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota Média")
plt.grid()
plt.show()
```
