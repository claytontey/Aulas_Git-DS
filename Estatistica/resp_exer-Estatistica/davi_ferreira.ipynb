## Exercício 20
Descreva e analise algo relevante nos dados.

```python
# Linguagem com maior tempo médio de resolução e linguagem com menor tempo médio:
medias = df.groupby("linguagem")["tempo_resolucao"].mean()

maior_ling = medias.idxmax()
menor_ling = medias.idxmin()

print(f"Linguagem com MAIOR tempo de resolução: {maior_ling} ({medias[maior_ling]:.2f})")
print(f"Linguagem com MENOR tempo de resolução: {menor_ling} ({medias[menor_ling]:.2f})")
```
