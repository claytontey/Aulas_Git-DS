## Exercício 12
Descreva e analise algo relevante nos dados.

```python
"""
Fora a média, analisar o desvio padrão dessas faltas é importante uma vez que 
ele representa o quao distante o valores estao de si, assim, dizendo se há uma 
uniformidade de presenca ou ausencia dos alunos
"""
vari = df["faltas"].var()
desv = sqrt(vari)
Coef_vari = desv/med_faltas
status_F="Equilibrado"

if (Coef_vari > 1.5):
    status_F = "Desequilibrado"
elif(Coef_vari < 0.6):
    status_F = "Moderadamente equilibrado"
elif(Coef_vari <0.3):
    status_F = "Muito equilibrado"
    
if(med_faltas > 5):
    status_F += " para maior proporcao de faltantes"
else:
    status_F += " para maior proporcao de presentes"

print(status_F)
```
