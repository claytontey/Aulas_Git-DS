```python
#voltando no acessos hora, que median e media mudou muito, vamos conferir quartis e outliers
print("Outliers:")
q1 = df['acessos_hora'].quantile(0.25)
q3 = df['acessos_hora'].quantile(0.75)
iqr = q3 - q1
l_superior = q3 + 1.5 * iqr
l_inferior = q1 - 1.5 * iqr
outliers1 = df[df['acessos_hora'] > l_superior]
outliers2= df[df['acessos_hora'] < l_inferior]
print(outliers1)
print(outliers2)
```
